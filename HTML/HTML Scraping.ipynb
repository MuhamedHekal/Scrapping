# HTML scraping 
- this project for scraping titles of movies, audience_score and number_of_audience_ratings from `Rotten Tomatoes` website for Top 100 movies 
- i will use BeautifulSoup lib for scrapping this data and store it in DataFrame 

### import libs that i need

```python
from bs4 import BeautifulSoup # lib for scraping 
import pandas as pd # lib for store the data in DataFrame
import os  # lib for read the file that conatian html pages 
```

```python
# first of all we need a list to append the data into it and then after all convert this list into DataFrame 
# and every raw of data we store it into dict into list 
data_list = []
folder_name = 'HtmlFile' # the folder that contain the html file 
for movie_html in os.listdir(folder_name): # iterate over the html pages 
    with open(os.path.join(folder_name,movie_html)) as file : ## we join the folder name and movieHtml name the get the all path not just the name of file 
        # Create a soup 
        soup = BeautifulSoup(file)
        # get the movie title 
        movie_title = soup.find('title').contents[0][:-len(' - Rotten Tomatoes')]
        # get the audience_score
        # the audience-score in the 'div' that its class name is audience-score meter
        audience_score = soup.find('div', class_ = 'audience-score meter').find('span').contents[0][:-1]
        # get the number_of_audience_ratings
        # it is in 'div' its class name is audience-info hidden-xs superPageFontColor
        number_of_audience_ratings = soup.find('div',class_ ='audience-info hidden-xs superPageFontColor')
        number_of_audience_ratings = number_of_audience_ratings.findAll('div')[1].contents[-1].strip().replace(',','')
        # append the movie_title ,  audience_score , number_of_audience_ratings in the data_list as a dictionary in each eteration
        data_list.append({
                            'title' : movie_title,
                            'audience_score' :audience_score ,
                            'number_of_audience_ratings' : number_of_audience_ratings})
```

```python
# convert the list to DataFrame 
df = pd.DataFrame(data_list , columns=['title' , 'audience_score','number_of_audience_ratings'])
df
```
